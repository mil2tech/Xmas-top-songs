After reviewing the data:

 - Drop columns 'url' , 'weekid'
 - need to determine what the column 'instance' means
 - In the field 'previous_week_position' , I will input the position 101 for nulls. Reasoning is the song was not in the top 100 songs list at the time. Consider the integer 101 to be the 'zero' for the data set for the field.
 - convert fields month day year into data time format

```python
# Make copy of original dataframe
charting_songs_copy = charting_songs.copy(deep=True)
```

```python
# Create variable to drop from dataframe, then execute removal
columns = ['url', 'weekid']
charting_songs_copy.drop(columns, axis=1, inplace=True)
```

```python
# Imput nulls with the integer 101.  101 will represent not being in the top 100 list the previous week before
charting_songs_copy['previous_week_position'] = charting_songs_copy['previous_week_position'].fillna(101)
```
